## **LIBRERIAS**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.max_columns = False
```

## **GR√ÅFICOS**

```python
tips = sns.load_dataset('tips')
```

### **Subplots - Columnas**

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='smoker')

plt.show()
```

### **Subplots - Filas**

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', row='smoker')

plt.show()
```

### **Subplots - Columnas y Filas**

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='smoker', row='time')

plt.show()
```

### **Subplots - Columnas agrupadas**

```python
sns.relplot(
    x='total_bill', 
    y='tip', 
    data=tips, 
    kind='scatter', 
    col='day', 
    col_wrap=2,
    col_order=['Thur', 'Fri', 'Sat', 'Sun'],
)

plt.show()
```

### **Customizing Scatterplots**

```python
sns.relplot(
    data=tips,
    x='total_bill',
    y='tip',
    kind='scatter',
    size='size',
    hue='smoker',
    style='smoker',
    alpha=0.9,
)

plt.show()
```

### **Line plots**

```python
df = pd.read_csv('https://raw.githubusercontent.com/datasets/covid-19/master/data/countries-aggregated.csv')
df = df[df['Country'].isin(['Colombia', 'Spain', 'Italy', 'France', 'Germany'])]
df['Date'] = pd.to_datetime(df['Date'])
df = df[df['Date'].dt.year == 2021]
```

```python
# Set the style like default

sns.relplot(
    data=df,
    x='Date',
    y='Confirmed',
    kind='line',
    hue='Country',
    style='Country'
)

plt.show()
```
