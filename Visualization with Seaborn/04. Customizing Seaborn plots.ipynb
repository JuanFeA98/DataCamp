## **Librerias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.max_columns = False
```

## **Gr√°ficos**

```python
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/mpg_ggplot2.csv')
```

### **1. Cambiando el color del fondo**

```python
# Puede tomar los siguientes valores: "white", "dark", "whitegrid", "darkgrid", "ticks"
sns.set_style('darkgrid')

sns.catplot(
    data=df,
    x='cyl',
    y='cty',
    kind='box',
)

plt.show()
```

### **2. Cambiando la paleta**

```python
sns.set_style('darkgrid')
sns.set_palette('RdBu')

sns.catplot(
    data=df,
    x='cyl',
    y='cty',
    kind='bar',
    ci=None
)

plt.show()
```

```python
sns.set_style('darkgrid')
sns.set_palette('Blues_r')

sns.catplot(
    data=df,
    x='cyl',
    y='cty',
    kind='bar',
    ci=None
)

plt.show()
```

#### **Usar una paleta personalizada**

```python
custom_palette = ['#98abc5', '#8a89a6', '#7b6888', '#6b486b']

sns.set_style('darkgrid')
sns.set_palette(custom_palette)

sns.catplot(
    data=df,
    x='cyl',
    y='cty',
    kind='bar',
    ci=None
)

plt.show()
```

### **3. Cambiando la escala**

```python
sns.set_style('darkgrid')
sns.set_palette('Blues_r')

# Puede tomar los siguientes valores: 'paper', 'notebook', 'talk', 'poster'
sns.set_context('talk')

sns.catplot(
    data=df,
    x='cyl',
    y='cty',
    kind='bar',
    ci=None
)

plt.show()
```

### **4. Agregando titulos**

```python
df = pd.read_csv('Datos/countries-of-the-world.csv')
```

```python
df['Birthrate'] = df['Birthrate'].str.replace(',', '.')
df['Birthrate'] = df['Birthrate'].apply(lambda x: float(x))
```

```python
g = sns.scatterplot(
    data=df, 
    x='Population',
    y='GDP ($ per capita)', 
    )
```

```python
g = sns.catplot(
    x='Region',
    y='Birthrate',
    data=df,
    kind='box'
)

g.fig.suptitle('New title', y=1.03)
plt.show()
```
