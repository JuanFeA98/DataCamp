## **Gráficos**

### **Gráfico de barras**

```python
df = pd.read_csv('Data/medals_by_country_2016.csv', index_col=0)
```

```python
df = df.sort_values('Gold', ascending=False)
```

```python
fig, ax = plt.subplots(figsize=(14, 5))

ax.bar(
    df.index, 
    df['Gold'],
    label='Gold'
)

ax.bar(
    df.index, 
    df['Silver'], 
    bottom=df['Gold'],
    label='Silver'
)

ax.bar(
    df.index, 
    df['Bronze'], 
    bottom=(df['Gold'] + df['Silver']),
    label='Bronze'
)

ax.set_xticks(df.index)
ax.set_xticklabels(df.index, rotation=45)

ax.set_ylabel('Número de medallas')
ax.legend()

plt.show()
```

### **Histogramas**

```python
df = pd.read_csv('Data/summer2016.csv', index_col=0)
```

```python
rowing = df.query('Sport == "Rowing"')
gymnastics = df.query('Sport == "Gymnastics"')
```

```python
fig, ax = plt.subplots()

ax.bar("Rowing", rowing['Height'].mean())
ax.bar("Gymnastics", gymnastics['Height'].mean())

ax.set_ylabel('Height (cm)')

plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(rowing['Height'], alpha=0.85, label='Rowing')
ax.hist(gymnastics['Height'], alpha=0.8, label='Gymnastics')

ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')

ax.legend()

plt.show()
```

### **Statistical Plotting**

```python
fig, ax = plt.subplots()

ax.bar("Rowing", rowing['Height'].mean(), yerr=rowing['Height'].std())
ax.bar("Gymnastics", gymnastics['Height'].mean(), yerr=gymnastics['Height'].std())

plt.show()
```

### **Boxplot**

```python
fig, ax = plt.subplots()

ax.boxplot([rowing['Height'], gymnastics['Height']])

ax.set_xticklabels(['Rowing', 'Gymnastics'])
ax.set_ylabel('Height (cm)')

plt.show()
```
