## **Gr√°ficos**

```python
seattle = pd.read_csv('Data/seattle_weather.csv')
austin = pd.read_csv('Data/austin_weather.csv')
```

```python
seattle_wf = seattle.groupby('DATE')[['MLY-TAVG-NORMAL', 'MLY-PRCP-NORMAL','MLY-PRCP-25PCTL', 'MLY-PRCP-75PCTL']].mean().reset_index()
austin_wf = austin.groupby('DATE')[['MLY-TAVG-NORMAL', 'MLY-PRCP-NORMAL','MLY-PRCP-25PCTL', 'MLY-PRCP-75PCTL']].mean().reset_index()
```

```python
meses = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
seattle_wf['MONTH'] = meses
austin_wf['MONTH'] = meses
```

```python
plt.style.use('ggplot')
fig, ax = plt.subplots(figsize=(7, 4))

ax.plot(austin_wf['MONTH'], austin_wf['MLY-TAVG-NORMAL'], marker='o', label='Austin')
ax.plot(seattle_wf['MONTH'], seattle_wf['MLY-TAVG-NORMAL'], marker='o', label='Seattle')

ax.set_xlabel('Time (months)')
ax.set_ylabel('Averarge temperature (Fahrenheit degrees)')

plt.legend()

plt.show()
```
