## **Librerias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.max_columns = False
```

## **Gr√°ficos**

```python
seattle = pd.read_csv('Data/seattle_weather.csv')
austin = pd.read_csv('Data/austin_weather.csv')
```

```python
seattle_wf = seattle.groupby('DATE')[['MLY-TAVG-NORMAL', 'MLY-PRCP-NORMAL','MLY-PRCP-25PCTL', 'MLY-PRCP-75PCTL']].mean().reset_index()
austin_wf = austin.groupby('DATE')[['MLY-TAVG-NORMAL', 'MLY-PRCP-NORMAL','MLY-PRCP-25PCTL', 'MLY-PRCP-75PCTL']].mean().reset_index()
```

```python
meses = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
seattle_wf['MONTH'] = meses
austin_wf['MONTH'] = meses
```

### **Simple plots**

```python
fig, ax = plt.subplots()

ax.plot(seattle_wf['MONTH'], seattle_wf['MLY-TAVG-NORMAL'])

plt.show()
```

### **Customizing your plots**

```python
fig, ax = plt.subplots()

ax.set_title('Weather in Seattle')
ax.set_xlabel('Time (Months)')
ax.set_ylabel('Average Temperature (Fahrenheit degrees)')

ax.plot(seattle_wf['MONTH'], seattle_wf['MLY-TAVG-NORMAL'], marker='o', color='#282c34')

plt.show()
```

### **Small Multiple Plots**

```python
fig, ax = plt.subplots(2, 1, figsize=(14, 6), sharey=True)

fig.patch.set_facecolor('#21252b') 

ax[0].plot(
    seattle_wf['MONTH'], seattle_wf['MLY-PRCP-NORMAL'], 
    marker='o', 
    color='orange')
ax[0].plot(seattle_wf['MONTH'], seattle_wf['MLY-PRCP-25PCTL'], 
           marker='o', 
           linestyle='--', 
           color='orange')
ax[0].plot(seattle_wf['MONTH'], seattle_wf['MLY-PRCP-75PCTL'], 
           marker='o', 
           linestyle='--', 
           color='orange')

ax[1].plot(
    austin_wf['MONTH'], austin_wf['MLY-PRCP-NORMAL'], 
    marker='o', 
    color='#287bff')
ax[1].plot(austin_wf['MONTH'], austin_wf['MLY-PRCP-25PCTL'], 
           marker='o', 
           linestyle='--', 
           color='#287bff')
ax[1].plot(austin_wf['MONTH'], austin_wf['MLY-PRCP-75PCTL'], 
           marker='o', 
           linestyle='--', 
           color='#287bff')


for i in range(2):
    ax[i].set_facecolor('#282c34') 

    ax[i].spines['top'].set_color('white')
    ax[i].spines['bottom'].set_color('white')
    ax[i].spines['left'].set_color('white')
    ax[i].spines['right'].set_color('white')

ax[0].set_ylabel('Precipitation (inches)', color='white', fontsize=13)
ax[1].set_ylabel('Precipitation (inches)', color='white', fontsize=13)
ax[1].set_xlabel('Time(months)', color='white', fontsize=13)

# ax[0].set_xticklabels(color='white')
# ax[1].set_yticklabels(list(np.range(12)), color='white')

# ax[0].set_yticklabels(color='white')

# ax[0].set_xticklabels([label_format.format(x) for x in meses])

ax[0].set_xticks(austin_wf['MONTH'])

plt.xticks(color='white')
plt.yticks(color='white')

plt.tick_params(axis='x', color='white')
plt.tick_params(axis='y', color='white')

plt.legend([1,2,3])

plt.show()
```

### **Final plot**

```python
fig, ax = plt.subplots(figsize=(14, 5))

ax.set_title('Average weather throughout the year', color='white', fontsize=15)

fig.patch.set_facecolor('#21252b') 
ax.set_facecolor('#282c34') 

ax.plot(seattle_wf['MONTH'], seattle_wf['MLY-TAVG-NORMAL'], marker='o')
ax.plot(austin_wf['MONTH'], austin_wf['MLY-TAVG-NORMAL'], marker='o')

ax.spines['top'].set_color('white')
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['right'].set_color('white')

plt.xticks(color='white')
plt.yticks(color='white')

plt.tick_params(axis='x', color='white')
plt.tick_params(axis='y', color='white')

plt.grid(linewidth=0.15)
plt.legend(['Austin weather', 'Seattle weather'], facecolor='#282c34', labelcolor='white', fontsize=10)

plt.show()
```
