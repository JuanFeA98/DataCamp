## **Librerias**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pd.options.display.max_columns = False
```

## **Gr√°ficos**

```python
seattle = pd.read_csv('Data/seattle_weather.csv')
austin = pd.read_csv('Data/austin_weather.csv')
```

```python
seattle_wf = seattle.groupby('DATE')[['MLY-TAVG-NORMAL']].mean().reset_index()
austin_wf = austin.groupby('DATE')[['MLY-TAVG-NORMAL']].mean().reset_index()
```

```python
meses = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC']
seattle_wf['MONTH'] = meses
austin_wf['MONTH'] = meses
```

### **Simple plots**

```python
fig, ax = plt.subplots()

ax.plot(seattle_wf['MONTH'], seattle_wf['MLY-TAVG-NORMAL'], marker='o')

plt.show()
```

### **Customizing your plots**

### **Final plot**

```python
fig, ax = plt.subplots(figsize=(14, 5))

plt.title('Average weather throughout the year', color='white', fontsize=15)

fig.patch.set_facecolor('#21252b') 
ax.set_facecolor('#282c34') 

ax.plot(seattle_wf['MONTH'], seattle_wf['MLY-TAVG-NORMAL'])
ax.scatter(seattle_wf['MONTH'], seattle_wf['MLY-TAVG-NORMAL'])

ax.plot(austin_wf['MONTH'], austin_wf['MLY-TAVG-NORMAL'])
ax.scatter(austin_wf['MONTH'], austin_wf['MLY-TAVG-NORMAL'])

ax.spines['top'].set_color('white')
ax.spines['bottom'].set_color('white')
ax.spines['left'].set_color('white')
ax.spines['right'].set_color('white')

plt.xticks(color='white')
plt.yticks(color='white')

plt.tick_params(axis='x', color='white')
plt.tick_params(axis='y', color='white')

plt.grid(linewidth=0.15)
plt.legend(['Austin weather', 'Seattle weather'], facecolor='#282c34', labelcolor='white', fontsize=10)

plt.show()
```
