## Missing Data

```python
df = pd.read_csv('Datos/diabetes.csv', index_col=0)
```

```python
df['triceps'].replace(0, np.nan, inplace=True)
df['insulin'].replace(0, np.nan, inplace=True)
df['bmi'].replace(0, np.nan, inplace=True)
```

```python
X_scaled = scale(df['glucose'])
```
