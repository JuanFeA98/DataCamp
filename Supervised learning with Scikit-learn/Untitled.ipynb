```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
```

```python
from sklearn import datasets
```

```python
plt.style.use('ggplot')
```

## Datos

```python
iris = datasets.load_iris()
```

```python
X = iris.data
y = iris.target
```

```python
df = pd.DataFrame(X, columns=iris.feature_names)
```

```python
df['target'] = np.where(df['target'] == 0, 
                    'setosa', 
                    np.where(df['target'] == 1,
                             'versicolor' , 
                             np.where(df['target'] == 2, 
                                      'virginica', 
                                      'Sin clasificar')))
```

```python
pd.plotting.scatter_matrix(
    df, 
    c=y, 
    figsize=[10,8],
    s=40,
    marker='o'
)
plt.show()
```

```python
plt.scatter(df['sepal length (cm)'], df['sepal width (cm)'], c=y)
plt.show()
```
